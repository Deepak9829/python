```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,2,3,4],[5,3,6,4],'r1',label='line1')
plt.plot([1,2,3,4],[5,3,6,4],linewidth=5,label='line2')
plt.plot([2,3,4,5],[4,5,6,7],'ro-',label='line3')
plt.plot([1,2,3,4],[5,6,4,3],color='red',linestyle='-',marker='o',markersize='5',markerfacecolor='g')
plt.axis([0,10,0,20])
plt.show()
t=np.arange(0,5,.2)
plt.plot(t,t,'r--',t,t,t**2,'bo',t,t**4,'c^')
```

```python
x=[10,50,40]
y=[20,56,20]
x1=[20,55,45]
y1=[40,50,15]
plt.plot(x,y,'g',linewidth=5,label='line1')
plt.plot(x1,y1,'r',linewidth=5,label='line2')
plt.legend()
plt.grid(color='k')
plt.title('my graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
